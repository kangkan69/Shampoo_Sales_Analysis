   Visualize how shampoo sales change over months to observe trends.

2. ðŸ“Š **Plot monthly and yearly patterns**  
   Use bar plots or line plots to see if sales vary by month or year.

3. ðŸ“‰ **Seasonal behavior analysis**  
   Analyze if there is a seasonal pattern in the sales data (e.g., summer vs winter).

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load dataset
df = pd.read_csv("cleaned_shampoo_sales.csv")


# Set style
sns.set(style="whitegrid")

# 1. Plot sales over time
plt.figure(figsize=(12, 6))
plt.plot(df['Month'], df['Sales'], marker='o')
plt.title("Shampoo Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Plot average sales by month
plt.figure(figsize=(10, 5))
sns.barplot(x='Month_Name', y='Sales', data=df, estimator='mean', order=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                                                                          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.title("Average Sales by Month")
plt.xlabel("Month")
plt.ylabel("Average Sales")
plt.show()

# 3. Plot total sales by year
plt.figure(figsize=(8, 5))
sns.barplot(x='Year', y='Sales', data=df, estimator='sum')
plt.title("Total Sales by Year")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.show()

# 4. Analyze seasonality
season_labels = {1: 'Winter', 2: 'Spring', 3: 'Summer', 4: 'Autumn'}
df['Season_Label'] = df['Season'].map(season_labels)

plt.figure(figsize=(8, 5))
sns.boxplot(x='Season_Label', y='Sales', data=df, order=['Winter', 'Spring', 'Summer', 'Autumn'])
plt.title("Sales Distribution by Season")
plt.xlabel("Season")
plt.ylabel("Sales")
plt.show()
```
